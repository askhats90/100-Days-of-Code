## Developing Number Guessing Game

```python
import random
```

```python
def make_guess():
    global num_of_attempts
    global guess
    if guess > num:
        num_of_attempts -= 1
        print("Too high.")
        if num_of_attempts > 0:
            print("Guess again.")
            print(f"You have {num_of_attempts} attempts remaining to guess the number.")
            print("---------------------------------------------------------------")
        else:
            print("You've run out of guesses, you lose.")
    elif guess < num:
        num_of_attempts -= 1
        print("Too low.")
        if num_of_attempts > 0:
            print("Guess again.")
            print(f"You have {num_of_attempts} attempts remaining to guess the number.")
            print("---------------------------------------------------------------")
        else:
            print("You've run out of guesses, you lose.")    
    else:
        num_of_attempts -= 1
        print("Bingo! You win.")
        print("---------------------------------------------------------------")
```

```python
while input("Do you want to play? y/n ") == 'y':
    num_list = list(range(1, 101))
    num = random.choice(num_list)
    print("---------------------------------------------------------------")
    print("Welcome to the Number Guessing Game!")
    print(f"I'm thinking of a number between {num_list[0]} and {num_list[-1]}.")
    difficulty = input("Choose a difficulty. Type 'easy' or 'hard': ")
    
    if difficulty == "easy":
        num_of_attempts = 10
    else:
        num_of_attempts = 5
    
    print("---------------------------------------------------------------")
    print(f"You have {num_of_attempts} attempts remaining to guess the number.")

    guess = 0
    while num_of_attempts > 0 and guess != num:
        guess = int(input("Make a guess: "))
        make_guess()
```
